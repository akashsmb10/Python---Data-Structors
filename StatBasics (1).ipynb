# THEORY

# 1. What is statistics, and why is it important?  
- Statistics is the science of collecting, organizing, analyzing, interpreting, and presenting data to make informed decisions.
- It is crucial for identifying patterns, testing hypotheses, and solving real-world problems in fields like economics, healthcare, and AI.

# 2. What are the two main types of statistics?  
-   
   1. **Descriptive Statistics**: Summarizes data (e.g., mean, median, charts).  
   2. **Inferential Statistics**: Draws conclusions about populations using samples (e.g., hypothesis testing).  

# 3. What are descriptive statistics?    
- Techniques to describe key features of datasets:  
   - Measures of central tendency (mean, median, mode)  
   - Measures of dispersion (variance, standard deviation)  
   - Visualization tools (histograms, pie charts).  

# 4. What is inferential statistics?   
- Methods to generalize from samples to populations, such as:  
     - Hypothesis testing  
     - Confidence intervals  
     - Regression analysis.  

# 5. What is sampling in statistics?   
- Selecting a subset (sample) from a population to estimate characteristics of the entire group.  
- Example: Surveying 1,000 voters to predict election results.  

# 6. What are the different types of sampling methods?  
-  **Random**:  
    - Simple random sampling  
    - Stratified sampling  
    - Cluster sampling  
- **Non-random**:  
    - Convenience sampling  
    - Purposive sampling  
    - Quota sampling.  

# 7. Difference between random and non-random sampling?  
- **Random**: Every member has an equal chance of selection (e.g., lottery draw).  
- **Non-random**: Subjective selection (e.g., surveying people in a mall).  

# 8. Define qualitative and quantitative data with examples.    
- **Qualitative**: Non-numerical (e.g., gender, feedback comments).  
- **Quantitative**: Numerical (e.g., height=180 cm, temperature=30°C).  

# 9. What are the different types of data in statistics?   
1. **Nominal**: Categories without order (e.g., blood types).  
2. **Ordinal**: Ordered categories (e.g., education levels).  
3. **Interval**: Equal intervals, no true zero (e.g., IQ scores).  
4. **Ratio**: True zero (e.g., weight in kg).  

# 11. What is the measure of central tendency?   
A single value representing the "center" of a dataset. Common measures:  
- Mean (average)  
- Median (middle value)  
- Mode (most frequent value).  

# 12. Define mean, median, and mode.  
- **Mean**: $\text{Mean} = \frac{\sum x_i}{N}$  
- **Median**: Middle value in sorted data.  
- **Mode**: Value appearing most frequently.  

# 13. What is the significance of the measure of central tendency?  
Provides a "typical" value for:  
- Comparing datasets  
- Simplifying complex data  
- Identifying data distribution patterns.  

# 14. What is variance, and how is it calculated?   
Average squared deviation from the mean:  
$\text{Variance} (\sigma^2) = \frac{\sum (x_i - \mu)^2}{N}$  

# 15. What is standard deviation, and why is it important?   
Square root of variance. Measures data spread in original units:  
$\text{SD} = \sqrt{\sigma^2}$.  
Key for understanding variability in datasets.  

# 16. Define and explain the term range in statistics.    
- Difference between maximum and minimum values:  
$\text{Range} = \text{Max} - \text{Min}$.  
Simple measure of data spread.  

# 17. What is the difference between variance and standard deviation?  
- **Variance**: Squared units (hard to interpret).  
- **Standard Deviation**: Original units (intuitive).  
Example: Variance = 25 kg² → SD = 5 kg.  

# 18. What is skewness in a dataset?  
Measure of asymmetry in data distribution:  
- **Positive skew**: Right tail longer (mean > median).  
- **Negative skew**: Left tail longer (mean < median).  

# 19. What does it mean if a dataset is positively or negatively skewed?  
- **Positive skew**: Most data concentrated on the left (e.g., income data).  
- **Negative skew**: Most data concentrated on the right (e.g., exam scores).  

# 20. Define and explain kurtosis.  
Measures "tailedness" of a distribution:  
- **High kurtosis**: Heavy tails (outliers likely).  
- **Low kurtosis**: Light tails (data clustered around mean).  

# 21. What is the purpose of covariance?    
Measures how two variables change together:  
- **Positive**: Variables increase/decrease together.  
- **Negative**: Variables move inversely.  

# 22. What does correlation measure in statistics?   
Standardized covariance (-1 to 1). Indicates:  
- **Strength**: How closely variables are related.  
- **Direction**: Positive or negative relationship.  

# 23. What is the difference between covariance and correlation?  
- **Covariance**: Scale-dependent (units vary).  
- **Correlation**: Scale-independent (dimensionless).  
Example: Correlation = 0.8 → Strong positive relationship.  

# 24. What are some real-world applications of statistics?    
- Healthcare: Drug efficacy analysis.  
- Finance: Stock market prediction.  
- Tech: A/B testing for app features.  
- Social Science: Census data analysis.

```python
# 1. Calculate mean, median, mode
import statistics

data = [12, 15, 18, 22, 25, 18, 20]
print(f"Mean: {statistics.mean(data)}")
print(f"Median: {statistics.median(data)}")
print(f"Mode: {statistics.mode(data)}")
```

```python
# 2. Python program for variance & standard deviation
def calculate_var_std(data):
    mean = sum(data)/len(data)
    variance = sum((x - mean)**2 for x in data) / len(data)
    std_dev = variance**0.5
    return variance, std_dev

data = [10, 20, 30, 40, 50]
var, std = calculate_var_std(data)
print(f"Variance: {var}\nStandard Deviation: {std}")
```

```python
# 4. Implement sampling techniques
import numpy as np

# Random sampling
population = np.arange(1, 101)
random_sample = np.random.choice(population, 10, replace=False)

# Stratified sampling (example)
strata = [np.arange(1, 21), np.arange(21, 41), np.arange(41, 61), np.arange(61, 81), np.arange(81, 101)]
stratified_sample = np.concatenate([np.random.choice(s, 2) for s in strata])

print(f"Random Sample: {random_sample}")
print(f"Stratified Sample: {stratified_sample}")
```

```python
# 5. Python function for range
def data_range(data):
    return max(data) - min(data)

data = [45, 32, 67, 21]
print(f"Range: {data_range(data)}")
```

```python
# 6. Plot histogram to visualize skewness
import matplotlib.pyplot as plt
import numpy as np

skewed_data = np.random.gamma(3, 2, 1000)
plt.hist(skewed_data, bins=30, edgecolor='black')
plt.title("Right-Skewed Distribution (Gamma)")
plt.show()
```

```python
# 7. Calculate skewness and kurtosis
from scipy import stats

print(f"Skewness: {stats.skew(skewed_data):.2f}")
print(f"Kurtosis: {stats.kurtosis(skewed_data):.2f}")
```

```python
# 8. Generate positive/negative skewed datasets
positive_skew = np.random.exponential(2, 1000)
negative_skew = np.random.beta(2, 5, 1000)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(positive_skew, bins=30, color='blue')
plt.title("Positive Skew")

plt.subplot(1, 2, 2)
plt.hist(negative_skew, bins=30, color='red')
plt.title("Negative Skew")
plt.show()
```

```python
# 9. Calculate covariance between two datasets
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
cov_matrix = np.cov(x, y)
print(f"Covariance: {cov_matrix[0, 1]}")
```

```python
# 10. Calculate correlation coefficient
corr = np.corrcoef(x, y)[0, 1]
print(f"Correlation Coefficient: {corr}")
```

```python
# 11. Scatter plot for two variables
plt.scatter(x, y, color='green', alpha=0.6)
plt.title("Scatter Plot of X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# 12. Compare random vs systematic sampling
systematic_sample = population[::10][:10]  # Interval k=10
print(f"Systematic Sample: {systematic_sample}")
print(f"Earlier Random Sample: {random_sample}")
```

```python
# 13. Mean/median/mode for grouped data
import pandas as pd

df = pd.DataFrame({
    "Midpoint": [15, 25, 35, 45],
    "Frequency": [5, 10, 8, 3]
})

grouped_mean = np.average(df["Midpoint"], weights=df["Frequency"])
print(f"Grouped Mean: {grouped_mean}")
```

```python
# 14. Simulate data and calculate stats
sim_data = np.random.normal(50, 10, 1000)
print(f"Mean: {np.mean(sim_data):.2f}")
print(f"SD: {np.std(sim_data):.2f}")
print(f"Range: {np.ptp(sim_data)}")
```

```python
# 15. Summarize data with pandas
df = pd.DataFrame(sim_data, columns=['Values'])
print(df.describe())
```

```python
# 17. Calculate IQR
Q3 = np.percentile(sim_data, 75)
Q1 = np.percentile(sim_data, 25)
iqr = Q3 - Q1
print(f"IQR: {iqr:.2f}")
```

```python
# 18. Z-score normalization
z_scores = (sim_data - np.mean(sim_data)) / np.std(sim_data)
plt.hist(z_scores, bins=30)
plt.title("Z-Score Normalized Data")
plt.show()
```

```python
# 20. Covariance heatmap
import seaborn as sns

data_matrix = np.random.randn(10, 5)
df = pd.DataFrame(data_matrix, columns=['A', 'B', 'C', 'D', 'E'])
sns.heatmap(df.cov(), annot=True)
plt.title("Covariance Heatmap")
plt.show()
```

```python
# 21. Correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# 22. Manual variance & SD
def manual_stats(data):
    mean = sum(data)/len(data)
    variance = sum((x - mean)**2 for x in data) / len(data)
    return variance, variance**0.5

data = [2, 4, 6, 8]
var, std = manual_stats(data)
print(f"Variance: {var}\nStandard Deviation: {std}")
```
