# Statistics Part 2 Theory

1. What is hypothesis testing in statistics?
 - Hypothesis testing is a statistical method used to make inferences about a population based on a sample of data. It involves formulating a null hypothesis and an alternative hypothesis and then using sample data to determine whether there is enough evidence to reject the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
 - The null hypothesis (H₀) is a statement that there is no significant difference between two populations or variables, or that there is no effect or relationship. The alternative hypothesis (H₁) is a statement that contradicts the null hypothesis and suggests there is a significant difference, effect, or relationship.

3. What is the significance level in hypothesis testing, and why is it important?
 - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type 1 error). It is typically set at 0.05 (or 5%). It is important because it determines the threshold for making a decision about rejecting the null hypothesis. A lower significance level requires stronger evidence to reject the null hypothesis.

4. What does a P-value represent in hypothesis testing?
 - The P-value is the probability of obtaining the observed results (or more extreme results) if the null hypothesis is true. It quantifies the evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
 - If the P-value is less than or equal to the significance level (P ≤ α), you reject the null hypothesis. This suggests that the observed results are unlikely to have occurred by random chance if the null hypothesis were true. If the P-value is greater than the significance level (P > α), you fail to reject the null hypothesis. This suggests that the observed results are reasonably likely to have occurred by random chance if the null hypothesis were true.

6. What are Type 1 and Type 2 errors in hypothesis testing?
 - Type 1 error (α): Rejecting the null hypothesis when it is actually true (also known as a false positive).
 - Type 2 error (β): Failing to reject the null hypothesis when it is actually false (also known as a false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 - **One-tailed test**: Used when the alternative hypothesis specifies a direction of the effect (e.g., the mean is greater than a certain value, or the mean is less than a certain value). The rejection region is in only one tail of the distribution.
 - **Two-tailed test:** Used when the alternative hypothesis does not specify a direction and simply states that there is a difference or relationship (e.g., the mean is different from a certain value). The rejection region is in both tails of the distribution.

8. What is the Z-test, and when is it used in hypothesis testing?
 - The Z-test is a statistical test used to compare a sample mean to a known population mean, or to compare two sample means, when the population standard deviation is known, or when the sample size is large (generally n > 30) so the sample standard deviation can be a good estimate of the population standard deviation. It assumes that the data is normally distributed.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

 - The Z-score measures how many standard deviations an observation or a sample mean is away from the mean of the population or the null hypothesis. The formula for a sample mean is:

 - $Z = (x̄ - μ) / (σ / √n)$

 - where:

 - x̄ is the sample mean
 - μ is the population mean (or the hypothesized population mean under the null hypothesis)
σ is the population standard deviation
n is the sample size
In hypothesis testing, the calculated Z-score is compared to a critical Z-value from the standard normal distribution to determine whether to reject the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
 - The T-distribution is a probability distribution that is similar in shape to the normal distribution but has heavier tails. It should be used instead of the normal distribution when the population standard deviation is unknown and the sample size is small (generally n < 30). As the sample size increases, the T-distribution approaches the normal distribution.

11. What is the difference between a Z-test and a T-test?
 - The main difference lies in the circumstances under which they are used and the distribution they rely on:

 - **Z-test:** Used when the population standard deviation is known or the sample size is large. Relies on the standard normal distribution.
 - **T-test:** Used when the population standard deviation is unknown and the sample size is small. Relies on the T-distribution.

12. What is the T-test, and how is it used in hypothesis testing?
 - The T-test is a statistical test used to compare sample means. It is used when the population standard deviation is unknown and the sample size is small. The T-test statistic is calculated similarly to the Z-score but uses the sample standard deviation as an estimate of the population standard deviation. The calculated T-statistic is then compared to a critical T-value from the T-distribution (with degrees of freedom equal to n-1) to make a decision about the null hypothesis.

13. What is the relationship between Z-test and T-test in hypothesis testing?
 - The T-test can be considered a generalization of the Z-test for situations where the population standard deviation is unknown and the sample size is small. As the sample size increases, the T-distribution approaches the normal distribution, and the results of a T-test become very similar to those of a Z-test.

14. What is a confidence interval, and how is it used to interpret statistical results?
 - A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It is constructed around a sample statistic (like the sample mean). For example, a 95% confidence interval means that if you were to repeat the sampling process many times, 95% of the calculated confidence intervals would contain the true population parameter. It helps interpret statistical results by providing a range of plausible values for the parameter, rather than just a single point estimate. If the confidence interval for a difference between two groups does not include zero, it suggests a statistically significant difference.

15. What is the margin of error, and how does it affect the confidence interval?
 - The margin of error is the range of values above and below the sample statistic that defines the confidence interval. It quantifies the uncertainty or variability in the estimate due to sampling. A larger margin of error results in a wider confidence interval, indicating greater uncertainty. The margin of error is influenced by the confidence level, the variability of the data (standard deviation), and the sample size (larger sample sizes lead to smaller margins of error).

16. How is Bayes' Theorem used in statistics, and what is its significance?
 - Bayes' Theorem is a fundamental theorem in probability theory that describes how to update the probability of a hypothesis based on new evidence. In statistics, particularly in Bayesian statistics, it is used to update prior beliefs about the probability of a hypothesis (the prior probability) based on observed data to obtain a posterior probability. Its significance lies in providing a framework for incorporating prior knowledge and updating beliefs as new data becomes available, which is useful in areas like medical diagnosis, machine learning, and risk assessment.

17. What is the Chi-square distribution, and when is it used?
 - The Chi-square (χ²) distribution is a skewed probability distribution that arises in statistical hypothesis testing, particularly when analyzing categorical data. It is used in tests that involve sums of squared differences, such as testing for independence between categorical variables (Chi-square test of independence) or comparing observed frequencies to expected frequencies (Chi-square goodness of fit test). The shape of the Chi-square distribution depends on its degrees of freedom.

18. What is the Chi-square goodness of fit test, and how is it applied?
 - The Chi-square goodness of fit test is used to determine whether an observed frequency distribution for a single categorical variable differs significantly from an expected frequency distribution. It is applied by calculating a Chi-square test statistic based on the differences between observed and expected frequencies, and then comparing this statistic to a critical value from the Chi-square distribution (with degrees of freedom equal to the number of categories minus 1). A significant Chi-square statistic suggests that the observed distribution does not fit the expected distribution.

19. What is the F-distribution, and when is it used in hypothesis testing?
 - The F-distribution is a probability distribution used in hypothesis testing, particularly when comparing variances or when performing analysis of variance (ANOVA). It is used in tests like ANOVA, which compare the means of three or more groups by examining the ratio of variance between groups to the variance within groups. The F-distribution is characterized by two degrees of freedom: one for the numerator and one for the denominator.

20. What is an ANOVA test, and what are its assumptions?
 - ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups. It tests whether there is a statistically significant difference between the means of the groups. The main assumptions of ANOVA are:
     - **Independence of observations:** The observations within each group and between groups are independent.
     - **Normality:** The data in each group are approximately normally distributed.
     - **Homogeneity of variances (homoscedasticity):** The variances of the populations from which the samples are drawn are equal.

21. What are the different types of ANOVA tests?

Some common types of ANOVA tests include:
   - **One-Way ANOVA:** Used to compare the means of three or more independent groups based on one categorical independent variable (factor).
   - **Two-Way ANOVA:** Used to examine the effect of two categorical independent variables on a continuous dependent variable, and to assess their interaction effect.
   - **Repeated Measures ANOVA:** Used when the same subjects are measured multiple times under different conditions or at different time points.

22. What is the F-test, and how does it relate to hypothesis testing?
 - The F-test is a statistical test that uses the F-distribution. It is primarily used to compare variances. In the context of ANOVA, the F-test is used to test the hypothesis that the means of multiple groups are equal. The F-statistic in ANOVA is the ratio of the variance between groups to the variance within groups. A larger F-statistic suggests greater variability between groups relative to within groups, providing evidence against the null hypothesis of equal group means. The calculated F-statistic is compared to a critical F-value from the F-distribution to determine whether to reject the null hypothesis.

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(k, p):
  """
  Calculates the Probability Distribution Function (PDF) for a Bernoulli distribution.

  Args:
    k: The outcome (0 or 1).
    p: The probability of success (value of k=1).

  Returns:
    The probability of the given outcome k.
  """
  if k == 1:
    return p
  elif k == 0:
    return 1 - p
  else:
    return 0 # For any other outcome
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm

# Example scenario: Test if the mean of a sample is significantly different from a known population mean.

# Known population parameters (under the null hypothesis)
population_mean_h0 = 75
population_std_dev = 10 # Assume population standard deviation is known

# Sample data
sample_data_ht = [78, 72, 80, 75, 79, 81, 73, 76, 77, 74]
sample_size_ht = len(sample_data_ht)
sample_mean_ht = np.mean(sample_data_ht)

# Significance level
alpha = 0.05

# Calculate the Z-statistic
z_statistic = (sample_mean_ht - population_mean_h0) / (population_std_dev / np.sqrt(sample_size_ht))

# Find the critical Z-value for a two-tailed test at the given alpha
# For alpha = 0.05, the critical values are approximately -1.96 and 1.96
critical_z_lower = norm.ppf(alpha / 2)
critical_z_upper = norm.ppf(1 - alpha / 2)

# Calculate the P-value (for a two-tailed test)
p_value_ht = 2 * (1 - norm.cdf(abs(z_statistic)))

print(f"Sample data: {sample_data_ht}")
print(f"Sample mean: {sample_mean_ht:.2f}")
print(f"Hypothesized population mean (H0): {population_mean_h0}")
print(f"Known population standard deviation: {population_std_dev}")
print(f"Sample size: {sample_size_ht}")
print(f"Calculated Z-statistic: {z_statistic:.2f}")
print(f"Critical Z-values (alpha={alpha}, two-tailed): {critical_z_lower:.2f} and {critical_z_upper:.2f}")
print(f"P-value: {p_value_ht:.4f}")

# Make a decision
if abs(z_statistic) > critical_z_upper: # Or p_value_ht <= alpha
  print("Decision: Reject the null hypothesis.")
  print("Conclusion: There is enough evidence to suggest the sample mean is significantly different from the population mean.")
else:
  print("Decision: Fail to reject the null hypothesis.")
  print("Conclusion: There is not enough evidence to suggest the sample mean is significantly different from the population mean.")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy.stats import t

# Example dataset (replace with your own dataset)
data_ci = [15, 18, 20, 22, 25, 28, 30, 32, 35]

# Confidence level
confidence_level = 0.95
alpha_ci = 1 - confidence_level

# Calculate sample mean, standard deviation, and size
sample_mean_ci = np.mean(data_ci)
sample_std_dev_ci = np.std(data_ci, ddof=1) # Use ddof=1 for sample standard deviation
sample_size_ci = len(data_ci)

# Calculate the degrees of freedom
degrees_of_freedom = sample_size_ci - 1

# Find the critical t-value for the desired confidence level and degrees of freedom
# For a two-sided interval, we use alpha/2 in each tail
critical_t_value = t.ppf(1 - alpha_ci / 2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error_ci = critical_t_value * (sample_std_dev_ci / np.sqrt(sample_size_ci))

# Calculate the confidence interval
confidence_interval_lower = sample_mean_ci - margin_of_error_ci
confidence_interval_upper = sample_mean_ci + margin_of_error_ci

print(f"Dataset: {data_ci}")
print(f"Sample mean: {sample_mean_ci:.2f}")
print(f"Sample standard deviation: {sample_std_dev_ci:.2f}")
print(f"Sample size: {sample_size_ci}")
print(f"Confidence level: {confidence_level}")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(f"Critical t-value: {critical_t_value:.2f}")
print(f"Margin of error: {margin_of_error_ci:.2f}")
print(f"Confidence Interval: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")

# Interpretation
print(f"\nInterpretation: We are {confidence_level * 100}% confident that the true population mean lies within the interval ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f}).")
```

```python
# 24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
  """
  Calculates the skewness and kurtosis of a dataset.

  Args:
    data: A list or NumPy array of numerical data.

  Returns:
    A tuple containing the skewness and kurtosis of the dataset.
  """
  # Calculate skewness
  skewness_value = skew(data)

  # Calculate kurtosis
  # By default, scipy.stats.kurtosis calculates Fisher's kurtosis,
  # where normal distribution has kurtosis of 0.
  kurtosis_value = kurtosis(data)

  return skewness_value, kurtosis_value

# Generate data from a normal distribution
mean_sk = 0
std_dev_sk = 1
sample_size_sk = 1000
normal_data_sk = np.random.normal(mean_sk, std_dev_sk, size=sample_size_sk)

# Calculate skewness and kurtosis for the normal data
skewness_norm, kurtosis_norm = calculate_skewness_kurtosis(normal_data_sk)

print(f"Dataset generated from Normal({mean_sk}, {std_dev_sk}) with size {sample_size_sk}")
print(f"Skewness: {skewness_norm:.4f}")
print(f"Kurtosis (Fisher): {kurtosis_norm:.4f}")

# For a perfectly normal distribution, skewness and kurtosis (Fisher) should be close to 0.
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy.stats import norm

# Example scenario: Test if the mean of a sample is significantly different from a known population mean.

# Known population parameters (under the null hypothesis)
population_mean_h0 = 75
population_std_dev = 10 # Assume population standard deviation is known

# Sample data
sample_data_ztest = [78, 72, 80, 75, 79, 81, 73, 76, 77, 74]
sample_size_ztest = len(sample_data_ztest)
sample_mean_ztest = np.mean(sample_data_ztest)

# Significance level
alpha_ztest = 0.05

# Calculate the Z-statistic
z_statistic_calculated = (sample_mean_ztest - population_mean_h0) / (population_std_dev / np.sqrt(sample_size_ztest))

# Find the critical Z-value for a two-tailed test at the given alpha
# For alpha = 0.05, the critical values are approximately -1.96 and 1.96
critical_z_lower_ztest = norm.ppf(alpha_ztest / 2)
critical_z_upper_ztest = norm.ppf(1 - alpha_ztest / 2)

# Calculate the P-value (for a two-tailed test)
p_value_ztest = 2 * (1 - norm.cdf(abs(z_statistic_calculated)))

print(f"Sample data: {sample_data_ztest}")
print(f"Sample mean: {sample_mean_ztest:.2f}")
print(f"Hypothesized population mean (H0): {population_mean_h0}")
print(f"Known population standard deviation: {population_std_dev}")
print(f"Sample size: {sample_size_ztest}")
print(f"Calculated Z-statistic: {z_statistic_calculated:.2f}")
print(f"Critical Z-values (alpha={alpha_ztest}, two-tailed): {critical_z_lower_ztest:.2f} and {critical_z_upper_ztest:.2f}")
print(f"P-value: {p_value_ztest:.4f}")

# Make a decision and interpret
if abs(z_statistic_calculated) > critical_z_upper_ztest: # Or p_value_ztest <= alpha_ztest
  print("\nDecision: Reject the null hypothesis.")
  print("Interpretation: There is statistically significant evidence to suggest that the sample mean is different from the hypothesized population mean.")
else:
  print("\nDecision: Fail to reject the null hypothesis.")
  print("Interpretation: There is not enough statistically significant evidence to suggest that the sample mean is different from the hypothesized population mean.")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
from scipy.stats import norm

def one_sample_ztest(sample_data, population_mean_h0, population_std_dev, alpha=0.05):
  """
  Performs a one-sample Z-test to compare a sample mean to a known population mean.

  Args:
    sample_data: A list or NumPy array of sample data.
    population_mean_h0: The hypothesized population mean under the null hypothesis.
    population_std_dev: The known population standard deviation.
    alpha: The significance level (default is 0.05).

  Returns:
    A tuple containing the Z-statistic, P-value, and the test decision.
  """
  sample_size = len(sample_data)
  sample_mean = np.mean(sample_data)

  # Calculate the Z-statistic
  z_statistic = (sample_mean - population_mean_h0) / (population_std_dev / np.sqrt(sample_size))

  # Calculate the P-value (two-tailed test)
  p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

  # Determine the decision
  if p_value <= alpha:
    decision = "Reject Null Hypothesis"
  else:
    decision = "Fail to Reject Null Hypothesis"

  return z_statistic, p_value, decision

# Example usage:
sample = [105, 108, 110, 102, 115, 112, 107, 109, 104, 111]
population_mean = 100
population_std = 5

z_stat, p_val, decision = one_sample_ztest(sample, population_mean, population_std)

print(f"Sample data: {sample}")
print(f"Hypothesized Population Mean: {population_mean}")
print(f"Known Population Standard Deviation: {population_std}")
print(f"Calculated Z-statistic: {z_stat:.4f}")
print(f"Calculated P-value: {p_val:.4f}")
print(f"Decision (at alpha=0.05): {decision}")

# Interpretation based on the decision
if decision == "Reject Null Hypothesis":
    print("Interpretation: There is statistically significant evidence to suggest the sample mean is different from the population mean.")
else:
    print("Interpretation: There is not enough statistically significant evidence to suggest the sample mean is different from the population mean.")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(population_mean_h0, population_std_dev, sample_size, alpha, true_population_mean=None):
  """
  Visualizes Type 1 and Type 2 errors for a one-sample Z-test scenario.

  Args:
    population_mean_h0: The hypothesized population mean under the null hypothesis.
    population_std_dev: The known population standard deviation.
    sample_size: The sample size.
    alpha: The significance level.
    true_population_mean: The actual population mean (for demonstrating Type 2 error).
                          If None, only Type 1 error is visualized.
  """
  # Calculate the standard error of the mean
  standard_error = population_std_dev / np.sqrt(sample_size)

  # Define the distribution under the null hypothesis (H0)
  mean_h0 = population_mean_h0
  std_dev_h0 = standard_error
  x_h0 = np.linspace(mean_h0 - 4 * std_dev_h0, mean_h0 + 4 * std_dev_h0, 200)
  pdf_h0 = norm.pdf(x_h0, mean_h0, std_dev_h0)

  # Find the critical values for the rejection region (two-tailed)
  # These are in terms of sample means
  critical_z_lower = norm.ppf(alpha / 2)
  critical_z_upper = norm.ppf(1 - alpha / 2)
  critical_mean_lower = mean_h0 + critical_z_lower * standard_error
  critical_mean_upper = mean_h0 + critical_z_upper * standard_error

  plt.figure(figsize=(12, 6))
  plt.plot(x_h0, pdf_h0, label='Distribution under H0')

  # Visualize Type 1 Error (Alpha)
  # Area under the H0 curve in the rejection regions
  x_reject_lower_h0 = x_h0[x_h0 <= critical_mean_lower]
  y_reject_lower_h0 = pdf_h0[x_h0 <= critical_mean_lower]
  plt.fill_between(x_reject_lower_h0, y_reject_lower_h0, color='red', alpha=0.3, label=f'Type 1 Error (α = {alpha})')

  x_reject_upper_h0 = x_h0[x_h0 >= critical_mean_upper]
  y_reject_upper_h0 = pdf_h0[x_h0 >= critical_mean_upper]
  plt.fill_between(x_reject_upper_h0, y_reject_upper_h0, color='red', alpha=0.3)

  plt.axvline(mean_h0, color='black', linestyle='--', label='Mean under H0')
  plt.axvline(critical_mean_lower, color='red', linestyle=':', label=f'Critical Mean Lower: {critical_mean_lower:.2f}')
  plt.axvline(critical_mean_upper, color='red', linestyle=':', label=f'Critical Mean Upper: {critical_mean_upper:.2f}')


  # Visualize Type 2 Error (Beta) if true_population_mean is provided
  if true_population_mean is not None:
    # Define the distribution under the alternative hypothesis (H1)
    mean_h1 = true_population_mean
    std_dev_h1 = standard_error # Assuming same standard deviation
    x_h1 = np.linspace(mean_h1 - 4 * std_dev_h1, mean_h1 + 4 * std_dev_h1, 200)
    pdf_h1 = norm.pdf(x_h1, mean_h1, std_dev_h1)
    plt.plot(x_h1, pdf_h1, label='Distribution under H1 (True Mean)')
    plt.axvline(mean_h1, color='green', linestyle='--', label=f'True Population Mean: {mean_h1}')

    # Visualize Type 2 Error (Beta)
    # Area under the H1 curve in the non-rejection region
    x_non_reject_h1 = x_h1[(x_h1 > critical_mean_lower) & (x_h1 < critical_mean_upper)]
    y_non_reject_h1 = pdf_h1[(x_h1 > critical_mean_lower) & (x_h1 < critical_mean_upper)]
    plt.fill_between(x_non_reject_h1, y_non_reject_h1, color='orange', alpha=0.5, label='Type 2 Error (β)')


  plt.xlabel('Sample Mean')
  plt.ylabel('Probability Density')
  plt.title('Visualization of Type 1 and Type 2 Errors')
  plt.legend()
  plt.grid(True, linestyle='--')
  plt.show()

# Example usage for visualizing Type 1 error:
print("Visualizing Type 1 Error:")
visualize_errors(population_mean_h0=100, population_std_dev=10, sample_size=30, alpha=0.05)

# Example usage for visualizing both Type 1 and Type 2 errors:
print("\nVisualizing Type 1 and Type 2 Errors (assuming true mean is 105):")
visualize_errors(population_mean_h0=100, population_std_dev=10, sample_size=30, alpha=0.05, true_population_mean=105)
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy.stats import norm, ttest_1samp

# --- Scenario: One-sample test comparing sample mean to a hypothesized value ---

# Define parameters
hypothesized_mean_compare = 60
true_population_mean_compare = 62 # Assuming a slightly different true mean
population_std_dev_compare = 10 # Known population standard deviation for Z-test

# Simulate data
sample_size_compare_small = 15 # Small sample size for T-test scenario
sample_size_compare_large = 100 # Large sample size for Z-test scenario

# Simulate a small sample
sample_data_small = np.random.normal(loc=true_population_mean_compare, scale=population_std_dev_compare, size=sample_size_compare_small)

# Simulate a large sample
sample_data_large = np.random.normal(loc=true_population_mean_compare, scale=population_std_dev_compare, size=sample_size_compare_large)

print(f"Hypothesized Mean (H0): {hypothesized_mean_compare}")
print(f"True Population Mean (for simulation): {true_population_mean_compare}")
print(f"Known Population Std Dev (for Z-test): {population_std_dev_compare}")

# --- Perform Z-test (using the large sample where Z-test is more appropriate) ---
print("\n--- Z-test (Large Sample) ---")
sample_mean_large = np.mean(sample_data_large)
sample_size_large = len(sample_data_large)

z_statistic_compare = (sample_mean_large - hypothesized_mean_compare) / (population_std_dev_compare / np.sqrt(sample_size_large))
p_value_z_compare = 2 * (1 - norm.cdf(abs(z_statistic_compare)))

print(f"Sample size: {sample_size_large}")
print(f"Sample Mean: {sample_mean_large:.2f}")
print(f"Z-statistic: {z_statistic_compare:.4f}")
print(f"P-value (Z-test): {p_value_z_compare:.4f}")
alpha_compare = 0.05
if p_value_z_compare <= alpha_compare:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")


# --- Perform T-test (using the small sample where T-test is more appropriate) ---
print("\n--- T-test (Small Sample, Population Std Dev Unknown) ---")
sample_mean_small = np.mean(sample_data_small)
sample_size_small = len(sample_data_small)

# We use the sample standard deviation for the T-test
# ttest_1samp handles calculating the sample std dev and degrees of freedom
t_statistic_compare, p_value_t_compare = ttest_1samp(sample_data_small, hypothesized_mean_compare)

print(f"Sample size: {sample_size_small}")
print(f"Sample Mean: {sample_mean_small:.2f}")
print(f"T-statistic: {t_statistic_compare:.4f}")
print(f"P-value (T-test): {p_value_t_compare:.4f}")
if p_value_t_compare <= alpha_compare:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")

# --- Comparison ---
print("\n--- Comparison ---")
print("Z-test is typically used when population standard deviation is known or sample size is large.")
print("T-test is typically used when population standard deviation is unknown and sample size is small.")
print("For large sample sizes, the T-distribution approaches the Normal distribution, and Z-test and T-test results will be very similar.")
print("For small sample sizes, the T-test accounts for the increased uncertainty due to estimating the population standard deviation from the sample, resulting in potentially larger P-values or wider confidence intervals compared to a Z-test (if the population std dev were known).")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import t

def calculate_confidence_interval(sample_data, confidence_level=0.95):
  """
  Calculates the confidence interval for a sample mean.

  Args:
    sample_data: A list or NumPy array of sample data.
    confidence_level: The desired confidence level (e.g., 0.95 for 95%).

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
  """
  sample_mean = np.mean(sample_data)
  sample_std_dev = np.std(sample_data, ddof=1) # Use ddof=1 for sample standard deviation
  sample_size = len(sample_data)

  # Calculate the degrees of freedom
  degrees_of_freedom = sample_size - 1

  # Calculate alpha
  alpha_ci_func = 1 - confidence_level

  # Find the critical t-value
  critical_t_value_func = t.ppf(1 - alpha_ci_func / 2, degrees_of_freedom)

  # Calculate the margin of error
  margin_of_error_func = critical_t_value_func * (sample_std_dev / np.sqrt(sample_size))

  # Calculate the confidence interval
  confidence_interval_lower_func = sample_mean - margin_of_error_func
  confidence_interval_upper_func = sample_mean + margin_of_error_func

  return confidence_interval_lower_func, confidence_interval_upper_func

# Example usage:
sample_for_ci = [15, 18, 20, 22, 25, 28, 30, 32, 35]
conf_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(sample_for_ci, conf_level)

print(f"Sample data: {sample_for_ci}")
print(f"Sample mean: {np.mean(sample_for_ci):.2f}")
print(f"Confidence Level: {conf_level * 100}%")
print(f"Calculated Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Significance Explanation
print("\nSignificance of the Confidence Interval:")
print(f"A {conf_level * 100}% confidence interval means that if we were to take many random samples from the same population and construct a confidence interval for each sample, approximately {conf_level * 100}% of those intervals would contain the true population mean.")
print(f"It provides a range of plausible values for the true population mean based on our sample data.")
print("A wider interval indicates more uncertainty about the true population mean (often due to smaller sample size or higher variability).")
print("A narrower interval indicates less uncertainty (often due to larger sample size or lower variability).")
```

```python
# 12.Perform a Chi-square test for independence between two categorical variables in Python
import numpy as np
from scipy.stats import chi2_contingency

# Example data: Contingency table (Observed frequencies)
# Let's say we're looking at the relationship between Gender (Male, Female)
# and Preference for a product (Product A, Product B).
# The rows represent one variable (e.g., Gender), and columns represent the other (e.g., Preference).

# Observed frequencies:
#              Product A   Product B   Total
# Male            50          30        80
# Female          40          70       110
# Total           90         100       190

observed_data = np.array([[50, 30],
                          [40, 70]])

# Perform the Chi-square test for independence
# chi2_contingency returns: chi2 statistic, p-value, degrees of freedom, expected frequencies
chi2_statistic_ind, p_value_ind_test, degrees_of_freedom_ind, expected_freq_ind = chi2_contingency(observed_data)

print("Observed Frequencies:")
print(observed_data)
print(f"\nCalculated Chi-square statistic: {chi2_statistic_ind:.4f}")
print(f"Calculated P-value: {p_value_ind_test:.4f}")
print(f"Degrees of freedom: {degrees_of_freedom_ind}")
print("Expected Frequencies:")
print(expected_freq_ind.round(2)) # Round for readability

# Interpretation (assuming alpha = 0.05)
alpha_ind_test = 0.05
if p_value_ind_test <= alpha_ind_test:
    print(f"\nDecision (at alpha={alpha_ind_test}): Reject the null hypothesis.")
    print("Interpretation: There is statistically significant evidence to suggest that Gender and Product Preference are dependent (not independent).")
else:
    print(f"\nDecision (at alpha={alpha_ind_test}): Fail to reject the null hypothesis.")
    print("Interpretation: There is not enough statistically significant evidence to suggest that Gender and Product Preference are dependent. We conclude they are independent.")
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Example Scenario: Rolling a six-sided die 120 times.
# We want to test if the die is fair.

# Observed frequencies (how many times each face landed)
# Let's say the outcomes were:
# Face 1: 18 times
# Face 2: 22 times
# Face 3: 20 times
# Face 4: 16 times
# Face 5: 25 times
# Face 6: 19 times
observed_frequencies_gof = np.array([18, 22, 20, 16, 25, 19])
total_observations = np.sum(observed_frequencies_gof)

# Expected frequencies (if the die is fair)
# For a fair six-sided die rolled 120 times, the expected frequency for each face
# is (Total Observations / Number of Categories) = 120 / 6 = 20
expected_frequencies_gof = np.array([20, 20, 20, 20, 20, 20])

print(f"Observed Frequencies: {observed_frequencies_gof}")
print(f"Expected Frequencies (under fair die assumption): {expected_frequencies_gof}")
print(f"Total Observations: {total_observations}")

# Perform the Chi-square goodness-of-fit test
# The chisquare function takes observed and expected frequencies.
# If expected is not provided, it assumes a uniform distribution.
chi2_statistic_gof, p_value_gof = chisquare(f_obs=observed_frequencies_gof, f_exp=expected_frequencies_gof)

print(f"\nCalculated Chi-square statistic (Goodness of Fit): {chi2_statistic_gof:.4f}")
print(f"Calculated P-value (Goodness of Fit): {p_value_gof:.4f}")

# Interpretation (assuming alpha = 0.05)
alpha_gof = 0.05
if p_value_gof <= alpha_gof:
    print(f"\nDecision (at alpha={alpha_gof}): Reject the null hypothesis.")
    print("Interpretation: The observed distribution is statistically significantly different from the expected distribution (i.e., the die is likely not fair).")
else:
    print(f"\nDecision (at alpha={alpha_gof}): Fail to reject the null hypothesis.")
    print("Interpretation: There is not enough statistically significant evidence to say the observed distribution is different from the expected distribution (i.e., the die appears to be fair).")
```
