# **# Restful API & Flask**

1. What is a RESTful API ?
 - REST API stands for REpresentational State Transfer API. It is a type of API (Application Programming Interface) that allows communication between different systems over the internet. REST APIs work by sending requests and receiving responses, typically in JSON format, between the client and server
 - REST APIs use HTTP methods (such as GET, POST, PUT, DELETE) to define actions that can be performed on resources. These methods align with CRUD (Create, Read, Update, Delete) operations, which are used to manipulate resources over the web.

2. Explain the concept of API specification.
 - API specification is a term that is often used interchangeably with API definition. While these terms have many similarities, they are different entities. An API specification provides a broad understanding of how an API behaves and how the API links with other APIs. It explains how the API functions and the results to expect when using the API. A good example of an API specification is the OpenAPI Specification. You can view the latest version of this specification (3.0.1) on GitHub.

3. What is Flask, and why is it popular for building APIs?
- Flask is a micro web framework written in Python that is used to build web applications and APIs. It is called a micro framework because it provides only the essential components needed to build web apps, leaving developers free to choose additional tools and libraries as needed.

- Flask is popular for building APIs because of the following reasons:

      - Lightweight and Simple: It has a minimalistic structure that makes it easy to learn and use, especially for beginners.
      - Flexible and Modular: Developers can design the application structure as they wish and use only the components they need.
      - Easy Routing: Flask allows developers to easily define API endpoints and handle different HTTP methods like GET, POST, etc.
      - Rich Extensions: Tools like Flask-RESTful and Flask-JWT simplify API creation, authentication, and database handling.
      - Good Community Support: Flask has excellent documentation and a large community, which helps developers solve problems quickly.
      - Flask is ideal for building RESTful APIs and small to medium-sized web applications.

4. What is routing in Flask?
- Routing in Flask refers to mapping URLs to specific functions. When a user accesses a particular URL, Flask uses the route to call the corresponding function. Each route is defined using the @app.route() decorator. This is how Flask determines what code to execute when a user visits a certain path.

6. What are HTTP methods used in RESTful APIs?
 - Common HTTP methods in RESTful APIs include:
         - GET: Retrieve data.
         - POST: Submit data to be processed.
         - PUT: Update existing data.
         - DELETE: Remove data.
         - PATCH: Partially update data.
 - These methods help perform CRUD operations.

7. What is the purpose of the @app.route() decorator in Flask?
 - The @app.route() decorator in Flask binds a URL to a view function. When a user accesses that URL, Flask runs the associated function and returns a response. It is essential for handling client requests in Flask applications.

8. What is the difference between GET and POST HTTP methods?
 - GET requests data from a server and displays it. Data is sent in the URL.
 - POST submits data to the server for processing (e.g., form data). Data is sent in the request body.
GET is used for retrieving, POST for sending data securely.



9. How do you handle errors in Flask APIs?

```python
@app.errorhandler(404)
def not_found(error):
    return {'error': 'Not found'}, 404
# This allows custom messages and HTTP status codes for errors like 404 or 500.
```

```python
# 9. How do you redirect to a different route in Flask?
from flask import redirect, url_for, Flask, render_template

# Re-declare app and the index route as they were defined in previous cells
# This is necessary because the notebook cells are executed sequentially and the variable 'app' might have been redefined or modified.
# The index route needs to be defined for url_for('index') to work.
app = Flask(
    __name__,
    static_folder='assets',  # Custom folder name instead of 'static'
    static_url_path='/public'  # Access files at /public instead of /static
)

# Re-define the route for '/' with the 'index' function
@app.route('/')
def index():
    return render_template('index.html')

# To demonstrate redirect, let's define a new route that redirects to 'index'
@app.route('/go-home')
def go_home():
    # Redirect to the 'index' endpoint
    return redirect(url_for('index'))

# To see this in action within the notebook, you can simulate a request to '/go-home'
with app.test_request_context('/go-home'):
    response = app.full_dispatch_request()
    # Print the status code and the location header of the redirect response
    print(f"Status Code: {response.status_code}")
    print(f"Location Header: {response.headers.get('Location')}")
```

```python
# 10. How do you handle errors in Flask (e.g., 404)?
from flask import Flask, render_template, abort

app = Flask(__name__)

# Custom error handlers
@app.errorhandler(404)
def page_not_found(error):
    return render_template('404.html'), 404

@app.errorhandler(500)
def internal_error(error):
    return render_template('500.html'), 500

# Example routes
@app.route('/')
def home():
    return "Welcome to the Home Page"

@app.route('/simulate-404')
def simulate_404():
    abort(404)

@app.route('/simulate-error')
def simulate_error():
    # This will trigger a 500 error
    result = 1 / 0
    return str(result)

if __name__ == '__main__':
    app.run(debug=False)  # Always disable debug in production!
```

```python
# 13. How can you redirect with query parameters in Flask?
from flask import Flask, redirect, url_for, request

app = Flask(__name__)

# Route that performs the redirect
@app.route('/old')
def old_route():
    # Redirect with multiple query parameters
    return redirect(url_for('new_route',
        category='books',
        page=2,
        sort='price'
    ))

# Route that receives the parameters
@app.route('/new')
def new_route():
    # Access query parameters
    category = request.args.get('category', 'all')
    page = request.args.get('page', 1, type=int)
    sort = request.args.get('sort', 'relevance')

    return f'''
    Category: {category}<br>
    Page: {page}<br>
    Sort: {sort}
    '''

# Dynamic parameter example
@app.route('/search')
def search():
    search_term = request.args.get('q', '')
    return redirect(url_for('results',
        q=search_term,
        filters='in_stock'
    ))

@app.route('/results')
def results():
    return f'''
    Search results for: {request.args.get('q')}<br>
    Filters: {request.args.get('filters')}
    '''

if __name__ == '__main__':
    app.run(debug=True)
```
